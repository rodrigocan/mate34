Baixando arquivo do corpus:

```python
import requests
url = 'https://raw.githubusercontent.com/rodrigocan/mate34/main/cojur.train.txt'
r = requests.get(url)
cojur = r.text
```

Primeiros testes com patterns:

```python
import re

patterns = ['LEI', 'Lei', 'lei', 'aaaaa']

for pattern in patterns:
    print('Procurando por "%s" em: \n"%s"' % (pattern, cojur[:100]))
    
    # Check pelo padrão
    if re.search(pattern, cojur):
        print('\n')
        print('Match foi encontrado.\n')
    else:
        print('\n')
        print('Sem Match encontrado.\n')
```

Encontrando todas as instâncias dos patterns:

```python
for pattern in patterns:
    print('Procurando por "%s" em: \n"%s"' % (pattern, cojur[:100]))

    matches = re.findall(pattern, cojur)
    
    # Check pelo padrão
    if matches:
        print(matches)
        print('\n')
    else:
        print('Sem matches.\n')
```

Mudando para case insensitive:

```python
pattern = 'lei'

print('Procurando por "%s" em: \n"%s"' % (pattern, cojur[:100]))

matches = re.findall(pattern, cojur, re.IGNORECASE)

if matches:
  print(matches)
else:
  print('Sem matches.')
```

Pattern para pegar as ocorrências de Lei no (número) de (dia) de (mês) de (ano):

```python
pattern = 'lei no \d+.\d+\D+\d+\D+\d+'

print('Procurando por leis relacionadas:')

matches = re.findall(pattern, cojur, re.IGNORECASE)

print(len(matches))

if matches:
  print(matches)
else:
  print('Sem leis relacionadas.')
```

Generalizando o pattern para pegar leis e decretos:

```python
pattern = '(?:lei|decreto) no \d+.\d+\D+\d+\D+\d+'

print('Procurando por leis relacionadas:')

matches = re.findall(pattern, cojur, re.IGNORECASE)

print(len(matches))

if matches:
  print(matches)
else:
  print('Sem leis relacionadas.')
```

Removendo parênteses:

```python
matches_sem_parenteses = [match.replace(")", "") for match in matches]
print(matches_sem_parenteses[:5])
```

Removendo espaços extras:

```python
clean_matches = [re.sub(' +', ' ', match).replace(' , ', ', ') for match in matches_sem_parenteses]
print(len(clean_matches))
print(clean_matches[:5])
```

Conjunto (set) dos matches, sem repetição:

```python
conjunto_matches = set(clean_matches)
print(len(conjunto_matches))
print(conjunto_matches)
```

Por ordem alfabética:

```python
conjunto_ordenado = sorted(conjunto_matches)

# Visualizando os 10 primeiros:
for norma in conjunto_ordenado[:10]:
  print(norma)

print('\n')

# Visualizando os 10 últimos:
for norma in conjunto_ordenado[-10:]:
  print(norma)
```

Separando leis e decretos em duas listas:

```python
leis = list(filter(lambda norma: norma.startswith(('LEI', 'Lei', 'lei')), conjunto_ordenado))
print(len(leis))
print(leis)

print('\n')

decretos = list(filter(lambda norma: norma.startswith(('DECRETO', 'Decreto', 'decreto')), conjunto_ordenado))
print(len(decretos))
print(decretos)
```

Visualizando os 20 primeiros e os vinte últimos elementos de cada lista:

```python
for lei in leis[:20]:
  print(lei)

print('\n')

for lei in leis[-20:]:
  print(lei)

print('\n')

for decreto in decretos[:20]:
  print(decreto)

print('\n')

for decreto in decretos[-20:]:
  print(decreto)
```

Melhorando o pattern e filtrando novamente as listas:

```python
pattern = re.compile('(?:lei|decreto) no \d+\.?\w+,\s+de\s+\d+\s+de\s+\D+\d+', re.IGNORECASE)

leis = [lei.capitalize() for lei in leis if pattern.match(lei)]

decretos = [decreto.capitalize() for decreto in decretos if pattern.match(decreto)]

for lei in leis[:20]:
  print(lei)

print('\n')

for lei in leis[-20:]:
  print(lei)

print('\n')

for decreto in decretos[:20]:
  print(decreto)

print('\n')

for decreto in decretos[-20:]:
  print(decreto)
```
